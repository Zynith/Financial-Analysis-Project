## <en><center>Intrinsic Value</center></en>

- Risk & Reward
- Leadership

<en><center>Always calculate the intrinsic value of a company before looking at the trading price.</center></en>

<en><center>Intrinsic value can be defined simply: It is the discounted value of the cash flows that can be taken out of the business during its life.</center></en>

<en><center>Price is what you pay. Value is what you get.</center></en>

### <en><center>Risks come from</center></en>

- Excessive Debt.
- Overpaying for an investment.
- Not knowing what you are doing.

#### <en><center>Debt/Equity Ratio</center></en>


- How much Debt for every Dollar of Equity.

- The __higher__ this is -> The more __risky__ the company appears.

- Industry dependent
    - Is it a manufacturing company that would require lots of machinery and equipment?
 
- Is it managed well?
    - Is the ratio gradually decreasing?
    - What is it from year to year?
    - Look at a period of about 10 years or more.
    - Is the company new? Did it need to buy equipment to get started?
    - A D/E ratio of 5 or more is very high.

Compound annual growth rate CAGR
- Fund managers often cite AAGR which gives extreme values

- However, we actually care about CAGR so we aren't tricked by fund managers.

```python
rev["5Y CAGR"] = (rev[2019]/rev[2014])** (1/5) -1
```

<en><center>Price - A fair Value</center></en> 

The 4 values we need are:
- Current EPS, Future EPS, Future P/E and Price

How we find the Price
- Current EPS -----> EPS Growth ----> Future EPS
- Future EPS ----> Future P/E
- Future P/E ----> Rate of Return ----> Price

Current EPS is reported
EPS growth rate: We won't use this one as its not accurate
BVPS growth rate - is better approximate for intrinsic value
Future P/E: min{2x Growth rate, avg.hist P/E}
- Be conservative take the lower number between {2x Growth rate, avg.hist P/E}

Return of investment: 15%

<en><center>P/E Ratio</center></en>

- P/E Ratio = price/EPS
- High P/E --> Business is expected to grow fast
- Low P/E --> business is expected to grow much
- Number is industry dependent
